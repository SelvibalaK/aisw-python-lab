```python
# basic if 
age = 18

if age >= 18:
    print("You are eligible to vote")
```

```python
# if / else 
age = 16

if age >= 18:
    print("You are eligible to vote")
else:
    print("You are not eligible to vote")
```

```python
# if / elif / else
marks = 72

if marks >= 90:
    grade = "A"
elif marks >= 75:
    grade = "B"
elif marks >= 60:
    grade = "C"
else:
    grade = "D"

print("Grade:", grade)
```

```python
# Conditions with Input (IMPORTANT)
user_age = int(input("Enter your age: "))

if user_age < 13:
    print("Child")
elif user_age < 20:
    print("Teenager")
elif user_age < 60:
    print("Adult")
else:
    print("Senior")
```

```python
# Logical Operators
username = "admin"
password = "1234"

if username == "admin" and password == "1234":
    print("Login successful")
else:
    print("Invalid credentials")
```

```python
# Common Mistake (INTENTIONAL)
num = 10

if num = 10:   # WRONG
    print("Equal")
```

```python
# The Fix
num = 10
if num == 10:
    print("Equal")
```
