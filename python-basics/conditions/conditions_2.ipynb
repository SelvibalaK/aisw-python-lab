```python
# basic if 
age = 18

if age >= 18:
    print("You are eligible to vote")
```

```python
# if / else 
age = 16

if age >= 18:
    print("You are eligible to vote")
else:
    print("You are not eligible to vote")
```

```python
# if / elif / else
marks = 72

if marks >= 90:
    grade = "A"
elif marks >= 75:
    grade = "B"
elif marks >= 60:
    grade = "C"
else:
    grade = "D"

print("Grade:", grade)
```

```python
# Conditions with Input (IMPORTANT)
user_age = int(input("Enter your age: "))

if user_age < 13:
    print("Child")
elif user_age < 20:
    print("Teenager")
elif user_age < 60:
    print("Adult")
else:
    print("Senior")
```

```python
# Logical Operators
username = "admin"
password = "1234"

if username == "admin" and password == "1234":
    print("Login successful")
else:
    print("Invalid credentials")
```

```python
# Common Mistake (INTENTIONAL)
num = 10

if num = 10:   # WRONG
    print("Equal")
```

```python
# The Fix
num = 10
if num == 10:
    print("Equal")
```

```python
# Nested Conditions ( Real Decision Flow ) example 1
age = int(input("Enter your age: "))

if age >= 18:
    has_id = input("Do you have a valid ID? (yes/no): ")
    if has_id == "yes":
        print("You are allowed to enter")
    else:
        print("ID required")
else:
    print("Age restriction applies")
```

```python
# Nested Conditions ( Real Decision Flow ) example 2
age = int(input("Enter your age: "))

if age >= 18:
    has_id = input("Do you have a valid ID? (yes/no): ")
    if has_id == "yes":
        print("You are allowed to enter")
    else:
        print("ID required")
else:
    print("Age restriction applies")

#case sensitive as it can't accept ' YES ' but accepts ' yes '
```

```python
# and – all conditions must be true

marks = 85
attendance = 80

if marks >= 75 and attendance >= 75:
    print("Eligible for exam")
else:
    print("Not eligible")
```

```python
# or – any one condition true

is_admin = False
is_teacher = True

if is_admin or is_teacher:
    print("Access granted")
else:
    print("Access denied")
```

```python
# not – reverse condition

logged_in = False

if not logged_in:
    print("Please log in first")
```

```python
# Membership Conditions ( in , not in )

allowed_users = ["admin", "manager", "staff"]
user = "admin"

if user in allowed_users:
    print("Access granted")
else:
    print("Access denied")
```

```python
# String-Based Conditions ( Case Handling )

choice = input("Enter yes or no: ").lower()

if choice == "yes":
    print("Confirmed")
elif choice == "no":
    print("Cancelled")
else:
    print("Invalid input")
```

```python
# Multiple Conditions Without elif ( Independent Checks )

num = 15

if num > 0:
    print("Positive")

if num % 5 == 0:
    print("Divisible by 5")

# Both can run if both are true
```

```python
# Short-hand if ( Ternary Operator ) example 1

age = 20
status = "Adult" if age >= 18 else "Minor"
print(status)
```

```python
# Short-hand if ( Ternary Operator ) example 2

num = 7
print("Even") if num % 2 == 0 else print("Odd")
```

```python
# Input Validation example 1

age = input("Enter age: ")

if age.isdigit():
    age = int(age)
    print("Valid age:", age)
else:
    print("Invalid input, numbers only")
```

```python
# Input Validation example 2

age = input("Enter age: ")

if age.isdigit():
    age = int(age)
    print("Valid age:", age)
else:
    print("Invalid input, numbers only")
```

```python
# Example 1 : Even or Odd

num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# Example 2 : Simple Login Check

username = input("Username: ")
password = input("Password: ")

if username == "admin" and password == "admin123":
    print("Login successful")
else:
    print("Invalid credentials")

# accepts only if the 
# Username : admin 
# Password : admin123
```
