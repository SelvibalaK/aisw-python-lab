```python
# Dictionary Operations
```

```python
student = {
    "name": " Park ",
    "age": 22
}
```

```python
# Add
student["city"] = " Seoul "
```

```python
# Update
student["age"] = 87
```

```python
# Delete
del student["city"]
```

```python
# Check key - whether the key is present in the dictionary
print("age" in student)
```

```python
print(student)
```

```python
# Looping Through Dictionaries
```

```python
student = {
    "name": " Park ",
    "age": 87,
    "course": "AI"
}
```

```python
# Keys
for key in student:
    print(key)
```

```python
# Values
for value in student.values():
    print(value)
```

```python
# Key + Value
for key, value in student.items():
    print(key, ":", value)
```

```python
# Login System

user = {
    "username": "admin",
    "password": "1234"
}

if user["username"] == "admin" and user["password"] == "1234":
    print("Login successful")
else:
    print("Invalid credentials")
```

```python
# Marks System

marks = {
    "maths": 85,
    "science": 78,
    "english": 90
}

total = sum(marks.values())
print("Total:", total)
print("Average:", total / len(marks))
```
