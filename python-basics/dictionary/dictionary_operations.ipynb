```python
# Dictionary Basics
```

```python
student = {
    "name": "Park",
    "age": 87,
    "course": "AI",
    "active": True
}

print(student)
print(student["name"])
print(student["course"])
```

```python
# Keys are unique
# Values can be anything ( int , string , bool , list , etc )
```

```python
# Dictionary Operations
```

```python
student = {
    "name": " Park ",
    "age": 22
}
```

```python
# Add
student["city"] = " Seoul "
```

```python
# Update
student["age"] = 87
```

```python
# Delete
del student["city"]
```

```python
# Check key - whether the key is present in the dictionary
print("age" in student)
```
