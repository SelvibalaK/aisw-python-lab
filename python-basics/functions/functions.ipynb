```python
# Basic function ( no parameters )

def greet():
    print("Hello, welcome to Python")

greet()
```

```python
# Simple add function with no return

def add(a, b):
    print(a + b)

add(10, 5)
add(20, 7)
```

```python
# Function with input

def check_age():
    age = int(input("Enter age: "))

    if age >= 18:
        print("Eligible")
    else:
        print("Not eligible")

check_age()
```
