```python
# Basic function ( no parameters )

def greet():
    print("Hello, welcome to Python")

greet()
```

```python
# Simple add function with no return

def add(a, b):
    print(a + b)

add(10, 5)
add(20, 7)
```

```python
# Function with input

def check_age():
    age = int(input("Enter age: "))

    if age >= 18:
        print("Eligible")
    else:
        print("Not eligible")

check_age()
```

```python
# Function with parameters

def greet(name):
    print("Hello", name)

greet("Selvibala")
```

```python
#Multiple parameters

def add(a, b):
    return a + b

print(add(3, 5))
```

```python
# Difference Between Return and Print

def square(n):
    return n * n

result = square(4)
print(result)
```

```python
# Default Parameters

def greet(name="User"):
    print("Hello", name)

greet()
greet("Selvibala")
```

```python
# Keyword Arguments

def student(name, age):
    print(name, age)

student(age=87, name="Alex")
```

```python
# Returning Multiple Values

def calculate(a, b):
    return a+b, a-b, a*b

add, sub, mul = calculate(10, 5)
print(add, sub, mul)
```
