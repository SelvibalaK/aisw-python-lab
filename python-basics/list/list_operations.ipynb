```python
# List Basics
```

```python
# Creating a list

numbers = [1, 2, 3, 4, 5]
print(numbers)
```

```python
# Mixed data types

student = ["Lee ", 87, True]
print(student)
```

```python
# Indexing

fruits = ["apple", "banana", "mango"]
print(fruits[0])
print(fruits[-1])
```

```python
# Slicing

print(fruits[0:2])
```

```python
# List operations and methods
```

```python
# used to add an element to the list (no preferred location)

numbers = [10, 20, 30]
numbers.append(40)
print(numbers)
```

```python
# used to insert an element to the list ( preferred location - here it is second position of the list )

numbers.insert(1, 15)
print(numbers)
```

```python
# to remove an certain element from the list 

numbers.remove(20)
print(numbers)
```

```python
# removes the last element of the list as of default 

numbers.pop()
print(numbers)
```

```python
# in ascending order

numbers.sort()
print(numbers)
```

```python
# in decesnding order

numbers.reverse()
print(numbers)
```
