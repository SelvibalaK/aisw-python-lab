```python
#basic loop 

for i in range(4):
    print("Hello")
```

```python
# loop syntax 
# for variable in sequence:
#    code
```

```python
# Print numbers

for i in range(5):
    print(i)
```

```python
# print 1 to 10

for i in range(1, 11):
    print(i)
```

```python
# Step value

for i in range(1, 11, 2):
    print(i)
```

```python
# Looping Through Strings

name = "Python"

for ch in name:
    print(ch)
```

```python
 # Looping Through Lists

fruits = ["apple", "banana", "mango"]

for fruit in fruits:
    print(fruit)
```

```python
# while Loop (Condition-based loop)
# syntax

# while condition:
#    code
```

```python
# Counting 
count = 1

while count <= 5:
    print(count)
    count += 1
```
