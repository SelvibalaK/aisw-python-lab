```python
# Set Operations
```

```python
skills = {" Python ", " AI "}
```

```python
# Add

skills.add(" Git ")
```

```python
# Add multiple

skills.update([" SQL ", " ML "])
```

```python
# Remove (error if not exists)

skills.remove(" AI ")
```

```python
# Remove (error if not exists)

skills.remove(" Math ")
```

```python
# Discard (no error)

skills.discard(" Cloud ")
```

```python
print(skills)
```

```python
# Set Mathematical Operations
```

```python
backend = {" Python ", " Django ", " SQL "}
frontend = {" Python ", " CSS ", " HTML "}
```

```python
# Union

print(backend | frontend)
```

```python
# Intersection

print(backend & frontend)
```

```python
# Difference

print(backend - frontend)
```
