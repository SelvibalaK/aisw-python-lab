```python
# Tuple Basics
```

```python
coordinates = (10, 20)
print(coordinates)
```

```python
# Accessing tuple values
print(coordinates[0])
```

```python
# Tuple unpacking
x, y = coordinates
print(x, y)
```

```python
# Trying to modify (will fail)
coordinates[0] = 50  # TypeError

# because tuples are immutable
```
