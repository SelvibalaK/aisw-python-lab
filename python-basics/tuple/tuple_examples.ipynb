```python
# Tuple Basics
```

```python
coordinates = (10, 20)
print(coordinates)
```

```python
# Accessing tuple values
print(coordinates[0])
```

```python
# Tuple unpacking
x, y = coordinates
print(x, y)
```

```python
# Trying to modify (will fail)
coordinates[0] = 50  # TypeError

# because tuples are immutable
```

```python
# Examples using Tuples
```

```python
location = (37.5186, 127.0267)

x, y = location
print("X:", x)
print("Y:", y)
```

```python
user_profile = ("Kim", 87, "AI")

print(user_profile[0])  # Name
print(user_profile[1])  # Age
```

```python
db_config = ("localhost", 5432)

print("Host:", db_config[0])
print("Port:", db_config[1])
```
