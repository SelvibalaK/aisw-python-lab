```python
# String to Integer
age_str = "87"
age = int(age_str)
print(age, type(age))
```

```python
# Integer to Float
count = 10
count_float = float(count)
print(count_float, type(count_float))
```

```python
# Float to Integer
height = 6.1
height_int = int(height)
print(height_int)
```
